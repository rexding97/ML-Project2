```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
athleteData = pd.read_csv('athleteData with GDP.csv')
athleteData = athleteData.round(4)
athleteData.head()
```

```python
athleteData.info()
```

```python
athleteData["Sex"] = athleteData["Sex"].replace({"M": 0, "F": 1})
```

```python
athleteData.head()
```

```python
regData = athleteData[["Sex", "Age", "Height", "Weight", "Sport", "Medal","GDPrate"]].copy()
regData["Medal"] = regData["Medal"].replace({"None": 0, "Gold": 1, "Silver": 1, "Bronze": 1})
```

```python
regData.fillna(0, inplace = True)
```

```python
regrouped = regData.groupby("Sport")
```

```python
def build(name, dataframe, weight = "balanced"):
    # Import RandomForestClassifier model from Scikit-Learn
    from sklearn.ensemble import RandomForestClassifier
    from sklearn.model_selection import train_test_split
    y = pd.DataFrame(dataframe['Medal'])
    X = pd.DataFrame(dataframe[["Sex", "Age", "Height", "Weight", "GDPrate"]]) 
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
    trainsample = len(X_train)/270521
    testsample = len(X_test)/270521
    # Create the Random Forest object
    rforest = RandomForestClassifier(n_estimators = 50,  # n_estimators denote number of trees
                                     max_depth = 4, class_weight = weight)       # set the maximum depth of each tr
  
    # Fit Random Forest on Train Data
    rforest.fit(X_train, y_train.values.ravel())
    # Import confusion_matrix from Scikit-Learn
    from sklearn.metrics import confusion_matrix
    # Predict Legendary values corresponding to Total
    y_train_pred = rforest.predict(X_train)  
    y_test_pred = rforest.predict(X_test)
    ratenomedal = len(y_test[y_test["Medal"] == 0])/len(X_test)
    
    cm = confusion_matrix(y_train, y_train_pred)
    if len(cm) == 2:
        TPR1 = cm[1][1]/(cm[1][1]+cm[1][0])
        FNR1 = cm[0][0]/(cm[0][0]+cm[0][1])
        PPV1 = cm[1][1]/(cm[1][1]+cm[0][1])
        FNV1 = cm[0][0]/(cm[0][0]+cm[1][0])
        predP1 = (cm[0][1]+cm[1][1])/len(X_train)
    else:
        TPR1 = np.NaN
        FNR1 = np.NaN
        PPV1 = np.NaN
        FNV1 = np.NaN
        predP1 = np.NaN
    cm = confusion_matrix(y_test, y_test_pred)
    if len(cm) == 2:
        TPR2 = cm[1][1]/(cm[1][1]+cm[1][0])
        FNR2 = cm[0][0]/(cm[0][0]+cm[0][1])
        PPV2 = cm[1][1]/(cm[1][1]+cm[0][1])
        FNV2 = cm[0][0]/(cm[0][0]+cm[1][0])
        predP2 = (cm[0][1]+cm[1][1])/len(X_train)
    else:
        TPR2 = np.NaN
        FNR2 = np.NaN
        PPV2 = np.NaN
        FNV2 = np.NaN
        predP2 = np.NaN
    return [name, trainsample, rforest.score(X_train, y_train),TPR1, FNR1, testsample, rforest.score(X_test, y_test), ratenomedal, TPR2, FNR2, PPV2, FNV2, predP2]

```

```python

acclst = []
for i, j in regrouped:
    acclst.append(build(i, j, None))
accdf = pd.DataFrame(acclst).rename(columns = {0:"Sport", 1: "TrainSample", 2:"TrainAcc", 3:"TPR1", 4: "FNR1", 5:"TestSample", 6:"TestAcc", 7:"RateNoMedal", 8:"TPR2", 9: "FNR2", 10:"PPV2", 11:"FOR2", 12:"RPP"})
accdf = accdf.set_index("Sport")
```

```python
accdf.dropna(inplace = True)
tmp = accdf.sort_values("TestAcc",ascending = False).head(10)
plt.figure(figsize=(16,8))
sb.heatmap(tmp, cmap = "YlGnBu", linewidth = 1, annot = True)
```

```python
accdf.sort_values("PPV2",inplace=True, ascending = False)
tmp = accdf.head(10)
plt.figure(figsize=(16,8))
sb.heatmap(tmp, cmap = "YlGnBu", linewidth = 1, annot = True)
```

```python
accdf.sort_values("RPP",inplace=True, ascending = False)
tmp = accdf.head(10)
plt.figure(figsize=(16,8))
sb.heatmap(tmp, cmap = "YlGnBu", linewidth = 1, annot = True)
```

```python
# Decision Tree using Train Data
from sklearn.tree import DecisionTreeClassifier
def dectree():
    dectree = DecisionTreeClassifier(max_depth = 3)  # create the decision tree object
    dectree.fit(X_train, y_train)                    # train the decision tree model
    y_test_pred = dectree.predict(X_test)
    score1 = dectree.score(X_train, y_train.values.ravel())
    score2 = dectree.score(X_test, y_test)
    return [score1, score2]


from sklearn.ensemble import RandomForestClassifier
def randomclass():
    rforest = RandomForestClassifier(n_estimators = 500,  # n_estimators denote number of trees
                                     max_depth = 4)       # set the maximum depth of each tree

    # Fit Random Forest on Train Data
    rforest.fit(X_train, y_train.values.ravel())
    y_train_pred = rforest.predict(X_train)
    y_test_pred = rforest.predict(X_test)
    score1 = rforest.score(X_train, y_train)
    score2 = rforest.score(X_test, y_test)
    return [score1, score2]


from sklearn.linear_model import LogisticRegression
def logreg():
    logreg = LogisticRegression(solver='liblinear')
    logreg.fit(X_train, y_train.values.ravel())
    logreg_prediction = logreg.predict(X_test)
    score1 = logreg.score(X_train, y_train)
    score2 = logreg.score(X_test, y_test)
    return [score1, score2]


from sklearn.naive_bayes import GaussianNB
def nb():
    nb = GaussianNB()
    nb.fit(X_train, y_train.values.ravel())
    nb_prediction = nb.predict(X_test)
    score1 = nb.score(X_train, y_train)
    score2 = nb.score(X_test, y_test)
    return [score1, score2]
```

```python
for i, j in regrouped:
    y = pd.DataFrame(j['Medal'])
    X = pd.DataFrame(j[["Sex", "Age", "Height", "Weight"]]) 

    from sklearn.model_selection import train_test_split

    acclst = []
    for i in range(100):
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
        d = dectree()
        r = randomclass()
        l = logreg()
        n = nb()
        tmplst = []
        tmplst.append(d[0])
        tmplst.append(r[0])
        tmplst.append(l[0])
        tmplst.append(n[0])
        tmplst.append(d[1])
        tmplst.append(r[1])
        tmplst.append(l[1])
        tmplst.append(n[1])
        acclst.append(tmplst)


    accdf = pd.DataFrame(acclst).rename(columns = {0:"dectree_train", 1: "randomclass_train", 2:"logreg_train", 3:"naivebayes_train", 
                                                   4:"dectree_test", 5: "randomclass_test", 6:"logreg_test", 7:"naivebayes_test"})
    print(accdf.describe())
    break
```
